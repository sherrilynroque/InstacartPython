### 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# get frequency of order_hour_of_day
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

#### Observations: there are 24 hours in a day divided by 3 categories (Most orders, Average orders, and Fewest orders). A block of 8 8 hours will be used in the for-loop condition. 

```python
# create a for-loop for the busiest hours of the day

hours = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12 ,16, 9]:
    hours.append("Most orders")
  elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
    hours.append("Average orders")
  else:
    hours.append("Fewest orders")
```

```python
# merge busiest hours of the day for-loops with ords_prods_merged dataframe
ords_prods_merged['busiest_hours_of_day'] = hours
```
