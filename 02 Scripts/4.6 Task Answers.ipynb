### df_orders_product_combined and df_prods_checked can be merged using the product_id field.

```python
# Drop unnamed column for df_prods_checked

df_prods_checked = df_prods_checked.drop(columns = ['Unnamed: 0'])
```

```python
# Check the output of df_products_checked

df_prods_checked.head()
```

```python
# Drop '_merge' column for df_orders_products_combined

df_orders_product_combined = df_orders_product_combined.drop(columns = ['_merge'])
```

```python
# Check the output of df_orders_product_combined

df_orders_product_combined.head()
```

```python
# Merge df_orders_product_combined with df_prods_checked using product_id as the key

df_orders_product_merged = df_orders_product_combined.merge(df_prods_checked, on = 'product_id', indicator = True)
```
